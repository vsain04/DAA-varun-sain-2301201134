```python
# Example 1: Find Fibonacci of 6
# Question: What is the 6th Fibonacci number?
def fib_recursive(n):
    if n <= 1:
        return n
    return fib_recursive(n-1) + fib_recursive(n-2)
print("Fibonacci Recursive (n=6):", fib_recursive(6))
```

```python
# Example 2: Find Fibonacci of 6 using Dynamic Programming
# Question: What is the 6th Fibonacci number (using dynamic programming)?
def fib_dp(n):
    if n <= 1:
        return n
    dp = [0, 1] + [0]*(n-1)
    for i in range(2, n+1):
        dp[i] = dp[i-1] + dp[i-2]
    return dp[n]
print("Fibonacci DP (n=6):", fib_dp(6))
```

```python
# Example 3: Sort the array [10, 7, 8, 9, 1, 5] using Merge Sort
# Question: Sort the array [10, 7, 8, 9, 1, 5] using merge sort.
def merge_sort(arr):
    if len(arr) > 1:
        mid = len(arr) // 2
        L = arr[:mid]
        R = arr[mid:]
        merge_sort(L)
        merge_sort(R)
        i = j = k = 0
        while i < len(L) and j < len(R):
            if L[i] < R[j]:
                arr[k] = L[i]
                i += 1
            else:
                arr[k] = R[j]
                j += 1
            k += 1
        while i < len(L):
            arr[k] = L[i]
            i += 1
            k += 1
        while j < len(R):
            arr[k] = R[j]
            j += 1
            k += 1
    return arr
example_arr = [10, 7, 8, 9, 1, 5]
print("Merge Sorted:", merge_sort(example_arr.copy()))
```

```python
# Example 4: Sort [10, 7, 8, 9, 1, 5] with Quick Sort
# Question: Sort the array [10, 7, 8, 9, 1, 5] using quick sort.
def quick_sort(arr):
    if len(arr) <= 1:
        return arr
    else:
        pivot = arr[0]
        less = [x for x in arr[1:] if x <= pivot]
        greater = [x for x in arr[1:] if x > pivot]
        return quick_sort(less) + [pivot] + quick_sort(greater)
print("Quick Sorted:", quick_sort(example_arr.copy()))
```

```python
# Example 5: Sort [10, 7, 8, 9, 1, 5] with Insertion Sort
# Question: Sort the array [10, 7, 8, 9, 1, 5] using insertion sort.
def insertion_sort(arr):
    for i in range(1, len(arr)):
        key = arr[i]
        j = i - 1
        while j >= 0 and arr[j] > key:
            arr[j + 1] = arr[j]
            j -= 1
        arr[j + 1] = key
    return arr
print("Insertion Sorted:", insertion_sort(example_arr.copy()))
```

```python
# Example 6: Sort with Bubble Sort
# Question: Sort the array [10, 7, 8, 9, 1, 5] using bubble sort.
def bubble_sort(arr):
    n = len(arr)
    for i in range(n):
        for j in range(0, n-i-1):
            if arr[j] > arr[j + 1]:
                arr[j], arr[j + 1] = arr[j + 1], arr[j]
    return arr
print("Bubble Sorted:", bubble_sort(example_arr.copy()))
```

```python
# Example 7: Sort using Selection Sort
# Question: Sort the array [10, 7, 8, 9, 1, 5] using selection sort.
def selection_sort(arr):
    n = len(arr)
    for i in range(n):
        min_idx = i
        for j in range(i + 1, n):
            if arr[j] < arr[min_idx]:
                min_idx = j
        arr[i], arr[min_idx] = arr[min_idx], arr[i]
    return arr
print("Selection Sorted:", selection_sort(example_arr.copy()))
```

```python
# Example 8: Binary Search for 8 in sorted array
# Question: Find the index of 8 in the sorted version of [10, 7, 8, 9, 1, 5].
def binary_search(arr, x):
    left, right = 0, len(arr) - 1
    while left <= right:
        mid = (left + right) // 2
        if arr[mid] == x:
            return mid
        elif arr[mid] < x:
            left = mid + 1
        else:
            right = mid - 1
    return -1
sorted_arr = sorted(example_arr)
print("Binary Search: Index of 8:", binary_search(sorted_arr, 8))
```
